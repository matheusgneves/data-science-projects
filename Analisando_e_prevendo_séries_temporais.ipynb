```python
from prophet import Prophet
```

```python
df_prophet = df_data[['data', 'contagem']].rename(columns={'data': 'ds', 'contagem': 'y'})
df_prophet.head()
```

```python
df_prophet = df_prophet.groupby('ds')['y'].sum().reset_index()
df_prophet
```

```python
import numpy as np

np.random.seed(4587)

modelo_0 = Prophet()
modelo_0.fit(df_prophet)

futuro_0 = modelo_0.make_future_dataframe(periods=90, freq='D')
previsoes_0 = modelo_0.predict(futuro_0)
```

```python
fig1 = modelo_0.plot(previsoes_0)
```

```python
previsoes_0[['ds', 'yhat', 'yhat_lower', 'yhat_upper']]
```

```python
fig2 = modelo_0.plot_components(previsoes_0)
```

```python
df_prophet.shape
```

```python
df_treino = pd.DataFrame()

df_treino['ds'] = df_prophet['ds'][:584]
df_treino['y'] = df_prophet['y'][:584]
```

```python
df_teste = pd.DataFrame()

df_teste['ds'] = df_prophet['ds'][584:]
df_teste['y'] = df_prophet['y'][584:]
```

```python
np.random.seed(4587)

modelo_1 = Prophet()
modelo_1.fit(df_treino)

futuro_1 = modelo_1.make_future_dataframe(periods=150, freq='D')
previsoes_1 = modelo_1.predict(futuro_1)
```

```python
fig1 = modelo_1.plot(previsoes_1)
plt.plot(df_teste['ds'], df_teste['y'], '.r')
```

```python
fig2 = modelo_1.plot_components(previsoes_1)
```

```python
df_previsao = previsoes_1[['ds', 'yhat']]
df_comparacao = pd.merge(df_previsao, df_teste, on='ds')
df_comparacao
```

```python
from sklearn.metrics import mean_squared_error
```

```python
mse = mean_squared_error(df_comparacao['y'], df_comparacao['yhat'])
rmse = np.sqrt(mse)
print(f'MSE: {mse}, RMSE: {rmse}')
```

```python
np.random.seed(4587)

modelo_1 = Prophet(yearly_seasonality=True)
modelo_1.fit(df_treino)

futuro_1 = modelo_1.make_future_dataframe(periods=150, freq='D')
previsoes_1 = modelo_1.predict(futuro_1)
```
